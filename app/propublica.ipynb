```python
import pandas as pd
import numpy as np
import requests
import json
import os
```

```python
propublica_token = os.environ['propublicatoken']
```

```python
root = 'https://api.propublica.org'

endpoint = '/congress/v1/117/house/members.json'

headers = {'X-API-Key': propublica_token}

r = requests.get(root + endpoint,
                headers = headers)
r
```

```python
myjson = json.loads(r.text)
housemembers = pd.json_normalize(myjson, record_path = ['results','members'])
housemembers
```
