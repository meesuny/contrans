```python
import pandas as pd
import numpy as np
import requests
import json
import os
```

```python
genius_token = os.environ['geniustoken']
```

```python
root = 'https://api.genius.com'
endpoint = '/search'

mypars = {'q': 'feelin good as hell', 'per_page': 21}

headers = {'Authorization': 'Bearer ' + genius_token}

r = requests.get(root + endpoint,
                 headers = headers,
                 params = mypars)  
r
```

```python
myjson = json.loads(r.text)
myjson['response']['hits'][1]['result']['full_title']
```

```python
myjson = json.loads(r.text)
genius_df = pd.json_normalize(myjson, record_path = ['response', 'hits'])
```
